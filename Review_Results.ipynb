```python
import pandas as pd
```

```python
df_random = pd.read_csv('output_random.csv')
```

```python
df_random
```

```python
sum(df_random[df_random['num_samples'] == 10000]['test_acc'])/3
```

```python
df_act_conf = pd.read_csv('output_active_confusion_using_loss.csv')
```

```python
df_act_conf = df_act_conf.dropna()
```

```python
df_act_conf
```

```python
sum(df_act_conf[df_act_conf['num_samples'] == 2000]['test_acc'])/2
```

```python
sample_count_list = [1000, 2000, 4000, 5000, 6000, 8000, 10000]

for sample_count in sample_count_list:
    rand_res = sum(df_random[df_random['num_samples'] == sample_count]['test_acc'])/3
    act_conf = sum(df_act_conf[df_act_conf['num_samples'] == sample_count]['test_acc'])/3

    print(f"Number of samples: {sample_count} | Random is: {rand_res} and Active Confusion is: {act_conf}")
```
